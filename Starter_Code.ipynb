## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'])
application_df.head()
```

```python
# Determine the number of unique values in each column.
unique_counts = application_df.nunique()
unique_counts
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
application_type_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
application_types_to_replace = application_type_counts[application_type_counts < 500].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app, "Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts[classification_counts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
classifications_to_replace = classification_counts[classification_counts < 100].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls, "Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df, drop_first=True)
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df["IS_SUCCESSFUL"].values
X = application_df.drop(columns=["IS_SUCCESSFUL"]).values

# Split the preprocessed data into a training and testing dataset
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Verify the number of features
input_features = X_train_scaled.shape[1]
print(f"Number of input features: {input_features}")

hidden_nodes_layer1 = 128  
hidden_nodes_layer2 = 64   
hidden_nodes_layer3 = 32   

nn = tf.keras.models.Sequential()

# First hidden layer with modified nodes
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=input_features))
nn.add(tf.keras.layers.BatchNormalization())
nn.add(tf.keras.layers.LeakyReLU())

# Dropout layer to mitigate overfitting
nn.add(tf.keras.layers.Dropout(0.5))

# Second hidden layer with modified nodes
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2))
nn.add(tf.keras.layers.BatchNormalization())
nn.add(tf.keras.layers.LeakyReLU())

# Dropout layer
nn.add(tf.keras.layers.Dropout(0.3))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3))
nn.add(tf.keras.layers.BatchNormalization())
nn.add(tf.keras.layers.LeakyReLU())

# Output layer (assuming a binary classification problem)
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
# Compile the model with a modified learning rate
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0005)  # Adjusted learning rate
nn.compile(loss="binary_crossentropy", optimizer=optimizer, metrics=["accuracy"])
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Define the early stopping criteria
early_stopping = EarlyStopping(monitor='val_loss', patience=25, verbose=1, restore_best_weights=True)

# Define model checkpoint to save the best model based on validation data
checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss', verbose=1, save_best_only=True, mode='min')

# Train the model
fit_model = nn.fit(
    X_train_scaled, y_train, 
    epochs=250, 
    validation_split=0.1, 
    batch_size=32, 
    callbacks=[early_stopping, checkpoint]
)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
import os

# Ensure the 'Models' directory exists, if not, create it
if not os.path.exists('Models'):
    os.makedirs('Models')

base_filename = "model_name.h5"
filename = os.path.join('Models', base_filename)
counter = 1

# While the filename exists, increment the counter and append to filename
while os.path.exists(filename):
    filename = os.path.join('Models', base_filename.replace('.h5', f'_{counter}.h5'))
    counter += 1

# Save the model to the determined filename within the 'Models' directory
nn.save(filename)

print(f"Model saved as {filename}")
```
